```python
# plot N_GC(Mass), z_step = 0.05
ntrees = 10 #100000
data   = np.zeros((ntrees,n_mass))
m_max  = 6e12 #[M_sun]
m_min  = 1e11 #[M_sun]
m_step = m_min
n_mass = int((m_max - m_min)/m_step)

data = np.zeros((ntrees,n_mass))

for i in range (0,ntrees):    
    N    = np.loadtxt(path2+'GCs-tree'+str(i+1)+'-func-of-mass.txt',usecols =(1))        
    data[i] = N
    
#mass is the same for all trees, so no need to put inside that loop    
mass = np.loadtxt(path2+'GCs-tree1-func-of-mass.txt',usecols =(0))    
#print(mass)

avg = np.zeros(n_mass)
for i in range(0,n_mass):
    avg[i] = np.mean(data[0:ntrees,i])

error = np.zeros(n_mass)
for i in range(0,n_mass):
    a0 = 0 
    for j in range(0,ntrees):
        a0 += (data[j,i] - avg[i])**2
    error[i] = np.sqrt(a0/((ntrees-1)*ntrees)) #standard error of the mean
    #error[i] = np.sqrt(a0/ntrees) #standard deviation

ax = plt.axes()
ax.set_xscale("log")
ax.set_yscale("log")
ax.set_xlabel("M(z=0) [M_sun]")
ax.set_ylabel("Number of GCs")
ax.errorbar(mass, avg,fmt='.', yerr = error)    
ax.set_title('10 trees')
```

```python
#plot the three different Pk of CDM, WDM
ntrees = 1000 
m_max  = 1e13 #[M_sun]
m_min  = 1e9  #[M_sun]
n_mass = 5
n_z    = 401-173 # nlev_max - z_form_GC_lev


data_M = np.zeros((ntrees,n_z))
data_C = np.zeros((ntrees,n_z))
data_CO = np.zeros((ntrees,n_z))

#redshift values are the same for all trees, so no need to put inside that loop    
z = np.loadtxt(path_z+'GCs-tree1M=1e9-func-of-z-COCO.txt',usecols =(0))    

for j in range (9,13+1):
    for i in range (0,ntrees):        
        data_M[i] = np.loadtxt(path_z+'GCs-tree'+str(i+1)+'M=1e'+str(j)+'-func-of-z-Mill.txt',usecols =(1)) 
        data_C[i] = np.loadtxt(path_z+'GCs-tree'+str(i+1)+'M=1e'+str(j)+'-func-of-z-CreaseyBBKS.txt',usecols =(1)) 
        data_CO[i] = np.loadtxt(path_z+'GCs-tree'+str(i+1)+'M=1e'+str(j)+'-func-of-z-COCO.txt',usecols =(1)) 

    avg_C = np.zeros(n_z)
    for i in range(0,n_z):
        avg_C[i] = np.mean(data_C[0:ntrees,i])

    error_C = np.zeros(n_z)
    for i in range(0,n_z):
        a0 = 0 
        for ii in range(0,ntrees):
            a0 += (data_C[ii,i] - avg_C[i])**2
        error_C[i] = np.sqrt(a0/ntrees) #standard deviation

    avg_CO = np.zeros(n_z)
    for i in range(0,n_z):
        avg_CO[i] = np.mean(data_CO[0:ntrees,i])

    error_CO = np.zeros(n_z)
    for i in range(0,n_z):
        a0 = 0 
        for ii in range(0,ntrees):
            a0 += (data_CO[ii,i] - avg_CO[i])**2
        error_CO[i] = np.sqrt(a0/ntrees) #standard deviation

    avg_M = np.zeros(n_z)
    for i in range(0,n_z):
        avg_M[i] = np.mean(data_M[0:ntrees,i])

    error_M = np.zeros(n_z)
    for i in range(0,n_z):
        a0 = 0 
        for ii in range(0,ntrees):
            a0 += (data_M[ii,i] - avg_M[i])**2
        #error3[i] = np.sqrt(a0/((ntrees-1)*ntrees)) #standard error of the mean
        error_M[i] = np.sqrt(a0/ntrees) #standard deviation
    plt.figure(figsize=[10,10])
    plt.semilogy(z, avg_C, '#3095'+str(j*(j-7)), label ='CDM Creasey+BBKS, M=1e'+str(j))
    plt.semilogy(z, avg_M, '#'+str(j*(j-7))+'30ff', label ='CDM Mill, M=1e'+str(j))
    plt.semilogy(z, avg_CO, '#ff'+str(j*(j-7))+'30', label ='WDM COCO, M=1e'+str(j))

    plt.fill_between(z, avg_CO-error_CO, avg_CO+error_CO,color='#ff'+str(j*(j-7))+'30',alpha = 0.1)
    plt.fill_between(z, avg_M-error_M, avg_M+error_M,color='#'+str(j*(j-7))+'30ff', alpha = 0.1)
    plt.fill_between(z, avg_C-error_C, avg_C+error_C,color='#3095'+str(j*(j-7)), alpha = 0.1)
    
    plt.xlim(8.65,20)
    #plt.ylim()
    plt.xlabel('z')
    plt.ylabel('N_GC')
    plt.title('1000 trees, M=1e'+str(j))
    
    plt.legend(loc=(1.1,0.0))
    plt.show()
```
